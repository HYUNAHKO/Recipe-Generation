```python
import os
import shutil

# 중첩된 폴더 경로 설정
nested_folder = "/home/elicer/yolov7/food_data_yai/images//Steak_grilled_beef"
parent_folder = "/home/elicer/yolov7/food_data_yai/images/train/Grilled_beef"

# 파일들을 중첩된 폴더에서 부모 폴더로 이동
def move_files_up(nested_dir, parent_dir):
    # 중첩된 폴더가 존재하는지 확인
    if not os.path.exists(nested_dir):
        print(f"[ERROR] The nested folder does not exist: {nested_dir}")
        return

    # 파일을 부모 폴더로 이동
    for file_name in os.listdir(nested_dir):
        source_path = os.path.join(nested_dir, file_name)
        target_path = os.path.join(parent_dir, file_name)

        try:
            shutil.move(source_path, target_path)
            print(f"[INFO] Moved {file_name} to {parent_dir}")
        except Exception as e:
            print(f"[ERROR] Failed to move {file_name}: {e}")

    # 중첩된 폴더가 비었으면 삭제
    try:
        os.rmdir(nested_dir)
        print(f"[INFO] Removed empty folder: {nested_dir}")
    except OSError as e:
        print(f"[ERROR] Failed to remove folder {nested_dir}: {e}")

# 함수 실행
move_files_up(nested_folder, parent_folder)
```

```python
import os
import shutil  # Import shutil for directory deletion

# 삭제할 파일들이 있는 디렉터리 경로
target_directory = "/home/elicer/yolov7/food_data_yai/labels/train/Bossam"

# 해당 경로에서 jpg 및 json 파일 삭제 및 폴더 삭제
def delete_jpg_and_json_files(directory):
    # 디렉터리 경로가 존재하는지 확인
    if not os.path.exists(directory):
        print(f"[ERROR] The directory does not exist: {directory}")
        return

    # 디렉터리 경로 출력
    print(f"[INFO] Target directory: {directory}")

    # .jpg 및 .json 확장자인 파일들 삭제
    for file_name in os.listdir(directory):
        # 파일 이름 출력
        print(f"[INFO] Checking file: {file_name}")

        # .jpg 또는 .txt 파일 삭제 조건 추가
        if file_name.endswith(".jpg") or file_name.endswith(".txt"):
            file_path = os.path.join(directory, file_name)
            try:
                os.remove(file_path)
                print(f"[INFO] Deleted file: {file_path}")
            except Exception as e:
                print(f"[ERROR] Failed to delete file {file_path}: {e}")

    # 빈 디렉터리 삭제
    try:
        os.rmdir(directory)
        print(f"[INFO] Deleted directory: {directory}")
    except OSError as e:
        print(f"[ERROR] Failed to delete directory {directory}: {e}")

# 삭제 함수 실행
delete_jpg_and_json_files(target_directory)
```

```python
import os
import shutil
import random

# 이미지 및 라벨 폴더 경로 설정
images_train_folder = "/home/elicer/yolov7/food_data_yai/images/train/Bossam"
labels_train_folder = "/home/elicer/yolov7/food_data_yai/labels/train/Bossam"
images_valid_folder = "/home/elicer/yolov7/food_data_yai/images/valid/Bossam"
labels_valid_folder = "/home/elicer/yolov7/food_data_yai/labels/valid/Bossam"

# train/valid 비율 설정 (예: 80% train, 20% valid)
train_ratio = 0.8

# 이미지 파일을 가져옴
image_files = [f for f in os.listdir(images_train_folder) if f.endswith(".jpg")]

# 이미지 파일을 무작위로 섞음
random.shuffle(image_files)

# train/valid 분할을 위해 인덱스 계산
train_size = int(len(image_files) * train_ratio)

# train/valid 분할
train_files = image_files[:train_size]
valid_files = image_files[train_size:]

# 유효성 검사 및 파일 이동 함수
def move_files(files, source_image_folder, source_label_folder, target_image_folder, target_label_folder):
    for file_name in files:
        # 이미지 파일 경로 설정
        image_source_path = os.path.join(source_image_folder, file_name)
        image_target_path = os.path.join(target_image_folder, file_name)

        # 라벨 파일 이름 설정 (확장자만 변경)
        label_file_name = file_name.replace(".jpg", ".txt")
        label_source_path = os.path.join(source_label_folder, label_file_name)
        label_target_path = os.path.join(target_label_folder, label_file_name)

        # 이미지 파일 및 라벨 파일이 모두 존재하는 경우에만 이동
        if os.path.exists(image_source_path) and os.path.exists(label_source_path):
            # 이미지 파일 이동
            shutil.move(image_source_path, image_target_path)
            # 라벨 파일 이동
            shutil.move(label_source_path, label_target_path)
            print(f"[INFO] Moved: {file_name} and corresponding label")
        else:
            # 라벨이 없으면 이미지 파일 삭제
            if os.path.exists(image_source_path):
                os.remove(image_source_path)
                print(f"[INFO] Deleted unmatched image file: {file_name}")

# 검증 세트(valid)로 파일 이동
move_files(valid_files, images_train_folder, labels_train_folder, images_valid_folder, labels_valid_folder)

print(f"[INFO] Train/Valid split complete. Train size: {len(train_files)}, Valid size: {len(valid_files)}")
```

```python
import os

# 대상 디렉터리 경로
target_directory = "/home/elicer/yolov7/food_data_yai/labels/train"
# 저장할 텍스트 파일 경로
output_file = "/home/elicer/yolov7/food_data_yai/ingredients_labels.txt"

def save_subdirectory_names_to_txt(directory, output_file):
    # 디렉터리 경로가 존재하는지 확인
    if not os.path.exists(directory):
        print(f"[ERROR] The directory does not exist: {directory}")
        return

    # 하위 폴더 이름 목록 초기화
    subdirectory_names = []

    # 하위 폴더 목록 가져오기
    for entry in os.listdir(directory):
        entry_path = os.path.join(directory, entry)
        if os.path.isdir(entry_path):
            subdirectory_names.append(entry)

    # 하위 폴더 이름을 텍스트 파일로 저장
    try:
        with open(output_file, 'w') as file:
            for name in subdirectory_names:
                file.write(name + '\n')
        print(f"[INFO] Subdirectory names saved to: {output_file}")
    except Exception as e:
        print(f"[ERROR] Failed to write to file {output_file}: {e}")

# 함수 실행
save_subdirectory_names_to_txt(target_directory, output_file)
```

```python
import os

def get_folder_size(folder_path):
    total_size = 0
    for dirpath, dirnames, filenames in os.walk(folder_path):
        for filename in filenames:
            file_path = os.path.join(dirpath, filename)
            total_size += os.path.getsize(file_path)
    return total_size

# Specify the main directory
main_directory = "/home/elicer/yolov7/food_data_yai/images/train"

# Iterate through each subfolder
for subfolder in os.listdir(main_directory):
    subfolder_path = os.path.join(main_directory, subfolder)
    if os.path.isdir(subfolder_path):
        folder_size = get_folder_size(subfolder_path)
        print(f"Size of '{subfolder}': {folder_size / (1024 * 1024):.2f} MB")
```

```python
import os

def create_folder_mapping(train_dir, valid_dir):
    folder_mapping = {}
    current_id = 0

    def add_folders_to_mapping(directory):
        nonlocal current_id
        for folder_name in os.listdir(directory):
            if not folder_name.startswith('.') and folder_name not in folder_mapping:
                folder_mapping[folder_name] = current_id
                current_id += 1

    add_folders_to_mapping(train_dir)
    add_folders_to_mapping(valid_dir)

    return folder_mapping

def update_class_ids(directory, folder_mapping):
    for folder_name in folder_mapping:
        folder_path = os.path.join(directory, folder_name)
        if os.path.exists(folder_path):
            class_id = folder_mapping[folder_name]
            print(f"Updating folder '{folder_name}' to class id '{class_id}'")
            for subdir, dirs, files in os.walk(folder_path):
                for file in files:
                    if file.endswith('.txt'):
                        file_path = os.path.join(subdir, file)
                        update_file_class_id(file_path, class_id)

def update_file_class_id(file_path, new_class_id):
    with open(file_path, 'r') as file:
        lines = file.readlines()

    # Filter out lines with negative numbers and remove duplicates
    filtered_lines = set()
    for line in lines:
        if not any(float(num) < 0 for num in line.split()[1:]):  # Skip lines with negative numbers
            new_line = f"{new_class_id} " + " ".join(line.split()[1:]) + '\n'
            filtered_lines.add(new_line)

    with open(file_path, 'w') as file:
        file.writelines(filtered_lines)

# Usage example
labels_root = '/home/elicer/yolov7/food_data_yai/labels'
train_dir = os.path.join(labels_root, 'train')
valid_dir = os.path.join(labels_root, 'valid')

folder_mapping = create_folder_mapping(train_dir, valid_dir)
update_class_ids(train_dir, folder_mapping)
update_class_ids(valid_dir, folder_mapping)
```
